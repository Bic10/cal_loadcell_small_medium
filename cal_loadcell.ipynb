```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress
```

```python
data = pd.read_csv('cal_loadcell_data_rp', sep = '\t', skiprows = [1])
data
```

```python
%matplotlib qt

plt.plot(data.vertical_load)
```

```python
horizontal_small_cycle1 = data.horizontal_load[3027:4906]
vertical_small_cycle1 = data.vertical_load[3027:4906]

horizontal_small_cycle2 = data.horizontal_load[6700:8713]
vertical_small_cycle2 = data.vertical_load[6700:8713]



fit_1 = linregress(horizontal_small_cycle1,vertical_small_cycle1)
slope1 = fit_1.slope
line1 = fit_1.slope*horizontal_small_cycle1+fit_1.intercept

fit_2 = linregress(horizontal_small_cycle2,vertical_small_cycle2)
slope2 = fit_2.slope
line2 = fit_2.slope*horizontal_small_cycle2+fit_2.intercept

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Primo subplot
ax1.scatter(horizontal_small_cycle1, vertical_small_cycle1, marker='o', s=20, color='blue', label='Data Points')
ax1.plot(horizontal_small_cycle1, line1, c='red', label=f'Fit Line (slope={fit_1.slope:.3f}, r={fit_1.rvalue:.3f})')
ax1.set_title('Cycle 1')
ax1.legend()
ax1.set_xlabel('Horizontal Load [V]')
ax1.set_ylabel('Vertical Load [kN]')

# Secondo subplot
ax2.scatter(horizontal_small_cycle2, vertical_small_cycle2, marker='o', s=20, color='green', label='Data Points')
ax2.plot(horizontal_small_cycle2, line2, c='red', label=f'Fit Line (slope={fit_2.slope:.3f}, r={fit_2.rvalue:.3f})')
ax2.set_title('Cycle 2')
ax2.legend()
ax2.set_xlabel('Horizontal Load [V]')
ax2.set_ylabel('Vertical Load [kN]')

# Miglioramento estetico
fig.tight_layout()

plt.show()
```

```python
slope = [slope1, slope2]

calibration_number_small = np.mean(slope)
```

```python
plt.plot(data.vertical_load)
plt.plot(data.horizontal_load*calibration_number_small)
```
